```python
for filename in os.listdir('features'):
  print(filename)
  setting = {x.split('=')[0]:x.split('=')[1] for x in filename.split('&')}
  Xs_u = np.load('features/' + filename)['Xs_u']
  num_pos_pos = np.sum(Xs_u[0,:,0,0])
  red, blue = float(setting['color_ratio']), 1-float(setting['color_ratio'])
  redpos, bluepos = num_pos_pos, float(setting['choice_edges']) - num_pos_pos
  importance = (redpos/red)/(bluepos/blue)
  print(red, blue, redpos,bluepos, importance, importance*24/(importance+1))
  
```
